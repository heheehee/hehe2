#### 多图布局

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def example_plot(ax,fontsize=12):
    ax.plot([1,2])
    ax.set_xlabel('x-label',fontsize=fontsize)
    ax.set_ylabel('y-label',fontsize=fontsize)
    ax.set_title('Title',fontsize=fontsize)
    
fig,axes=plt.subplots(2,2)
fig.set_facecolor("y")
example_plot(axes[0,0])
example_plot(axes[0,1])
example_plot(axes[1,0])
example_plot(axes[1,1])
# fig.tight_layout(w_pad=10,h_pad=10)#将图表重叠的地方整理出来
fig.subplots_adjust(0,0,1,1,wspace=1,hspace=0.8)
```

#### 散点图和直方图的合并实战

```python
atheletes=pd.read_csv("data/athlete_events.csv")[:4000]
male_atheletes=atheletes[atheletes['Sex']=='M'][['Height','Weight']]
widths=(4,1)
heights=(1,4)
fig=plt.figure(figsize=(10,10))
gs=fig.add_gridspec(2,2,width_ratios=widths,height_ratios=heights)
ax1=fig.add_subplot(gs[0,0])
ax1.hist(male_atheletes['Height'],bins=20)
for label in ax1.get_xticklabels():
    label.set_visible(False)

ax2=fig.add_subplot(gs[1,0])
ax2.scatter('Height','Weight',data=male_atheletes)

ax3=fig.add_subplot(gs[1,1])
ax3.hist(male_atheletes['Weight'],bins=20,orientation='horizontal')
ax3.set_yticks(range(0,160,120))
ax3.set_ylim(0,140)
for label in ax3.get_yticklabels():
    label.set_visible(False)
    
fig.tight_layout(h_pad=0,w_pad=0)
```

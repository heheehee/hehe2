#### 配置项

```python
from pyecharts.charts import Bar,Scatter
from pyecharts import options as opts
from faker import Faker
from pyecharts.commons.utils import JsCode
```

```python
c=(
    Bar(init_opts=opts.InitOpts(
        width="400px",
        height="300px",
        bg_color="#ccc",#更换背景颜色
        animation_opts=opts.AnimationOpts(animation=False)#初始化图像时是否有动画效果
    ))
    .add_xaxis(Faker.choose())
    .add_yaxis("商品",Faker.values())
)
c.render_notebook()
```

```python
c=(
    Bar(init_opts=opts.InitOpts(
        width="400px",
        height="300px"))
    .add_xaxis(Faker.choose())
    .add_yaxis("商品",Faker.values())
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="商品出售情况",
            subtitle="子标题",
            pos_right="20px",
            title_textstyle_opts=opts.TextStyleOpts(font_size=12)
        )
    )
    
)
c.render_notebook()
```

```python
c=(
    Bar(init_opts=opts.InitOpts(
        width="500px",
        height="300px"))
    .add_xaxis(Faker.clothes)
    .add_yaxis("第一季度",Faker.values(),gap="0%")
    .add_yaxis("第二季度",Faker.values(),gap="0%")
    .set_global_opts(
       legend_opts=opts.LegendOpts(selected_mode="single")
        
    )
    
)
c.render_notebook()
```

```python
y_values=[(x,y) for x,y in zip(Faker.values(),Faker.values())]
c=(
    Scatter(init_opts=opts.InitOpts(width="500px",height="300px"))
    .add_xaxis(Faker.clothes)
    .add_yaxis("销售",y_values)
    .set_global_opts(
         visualmap_opts=opts.VisualMapOpts(type_="color",range_text=["大","小"])
    )
)
c.render_notebook()
```

```python
c=(
    Scatter()
    .add_xaxis(Faker.clothes)
    .add_yaxis("商家A",Faker.values())
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Scatter-多维度数据"),
        tooltip_opts=opts.TooltipOpts(
#             formatter="{b}"
            formatter=JsCode("function(params){return params.value}")
        )
    )
)
c.render_notebook()
```

```python
athletes=pd.read_csv("data/athlete_events.csv")
athletes.head()
```

```python
countries = {
    'CHN':'中国',
    'JPN':"日本",
    'KOR':'韩国',
    'USA':"美国",
    'CAN':"加拿大",
    'BRA':"巴西",
    'GBR':"英国",
    'FRA':"法国",
    'ITA':"意大利",
    'ETH':"埃塞俄比亚",
    'KEN':"肯尼亚",
    'NIG':"尼日利亚",
}
my_athletes=athletes[athletes['NOC'].isin(list(countries.keys()))]
man_athletes=[]
woman_athelets=[]
for country in countries.keys():
    man_heights=my_athletes[(my_athletes['NOC']==country)&(my_athletes['Sex']=='M')]['Height'].dropna()
    woman_heights=my_athletes[(my_athletes['NOC']==country)&(my_athletes['Sex']=='F')]['Height'].dropna()
    man_athletes.append(man_heights)
    woman_athelets.append(woman_heights)
    
c=Boxplot(init_opts=opts.InitOpts(
    width="800px",
    height="400px",
    animation_opts=opts.AnimationOpts(animation=False)
))
c.add_xaxis(list(countries.values()))
c.add_yaxis("男性运动员",c.prepare_data(man_athletes))
c.add_yaxis("女性运动员",c.prepare_data(woman_athelets))
c.set_global_opts(
    yaxis_opts=opts.AxisOpts(interval=20,min_=120,max_=240)
)
c.render_notebook()
```

#### 地理图

```python
from pyecharts.charts import Map
from pyecharts import options as opts
from faker import Faker
```

```python
c=(
    Map()
    .add("淘宝商家商品售卖情况",[(x,y) for x,y in zip(Faker.provinces,Faker.values())],"china")
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts()
    )
)
c.render_notebook()
```

```python
c=(
    Map()
    .add("广东省各城市人口密度",[(x,y) for x,y in zip(Faker.guangdong_city,Faker.values())],"广东")
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts()
    )
)
c.render_notebook()
```

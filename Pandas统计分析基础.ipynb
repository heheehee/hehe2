# 掌握DataFrame的常用操作

```python
import pandas as pd
```

```python
#Series系列
ser1=pd.Series([1,2,'a'],index=['a','b','c'])
print(ser1)
ser2=pd.Series({'a':[1,2,3],'b':['1','2','3']})
print(ser2)
```

```python
#构建数据框(DataFrame)
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
print(d)
# df=pd.DataFrame(d)
df=pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)

#用字典构建数据框
d={'color':['blue','green','yellow','red','white'],
   'object':['ball','pen','pencil','paper','mug'],
  'price':[1.2,1.0,0.6,0.9,1.7]}
frame=pd.DataFrame(d,index=['a','b','c','d','e'])
print(frame)

print(pd.DataFrame(index=[1,2],columns=[1,2]))  #构建全是缺失值的数据框
print(pd.DataFrame(0,index=[1,2],columns=[1,2])) #构建数值全为0的数据框
```

```python
#数据框的常用属性
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df=pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
df.values #访访问数据框的值，以数组形式
print(df.index)
print(df.shape)
print(df.dtypes)
```

```python
#数据框的增删改查操作
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df=pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)

print(df['A']) #单列数据访问
print(df[['A','C']])  #多列数据的访问

print(df.head(3)) #访问数据的前三行数据
print(df.tail(3))

print(df.iloc[0,0]) #按照行列顺序进行数据访问
print(df.iloc[0:3,0]) #取到第一列前三行数据，切片操作
print(df.iloc[:,0]) #访问单列数据
print(df.iloc[0,:]) #访问单行数据
print(df.iloc[1:3,1:3]) #取出目标元素的数据框

#通过行列名称进行数据访问
print(df.loc['a','A'])
print(df.loc['a':'c','A']) #左闭右闭区间
print(df.loc[:,'A']) #访问某一列数据
print(df.loc['a',:]) #访问某一行数据
print(df.loc[['b','c'],['B','C']]) #取出目标元素的数据框
 #返回数据框(注意如下方式返回值的区别)
print(df.iloc[:,0])
print(df.iloc[:,0:1])
print(type(df.iloc[:,0:1]))
```

```python
#修改数据框中的元素
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df=pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
df.loc['a','A']=101
df.loc[:,'B']=0.25
df.loc[:,'C']=[1,2,3,4]
print(df)
```

```python
#为数据框增添数据
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df=pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
df['E']=5
df['F']=[1,2,3,4]
print(df)
```

```python
#删除数据框中的元素
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df=pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
print(df.drop('D',axis=1,inplace=False))
print(df.drop(['a','c'],axis=0))
```

# 描述分析DataFrame数据

```python
import numpy as np
d=[[1.3,2.0,3,4],[2,4,1,4],[2,5,1.9,7],[3,1,0,11]]
df=pd.DataFrame(d,index=['a','b','c','d'],columns=['A','B','C','D'])
print(df)
print(np.mean(df,axis=1)) #每一列元素执行
print(df.mean(axis=1))
print(df.std())
print(df.describe())
```

```python
import pandas as pd
```

```python
order['lock_time']
```

```python
print(pd.DatetimeIndex(order['lock_time']))
print(pd.PeriodIndex(order['lock_time'],freq='H'))
```

```python
print(order['lock_time']+pd.Timedelta(days=1))
print(order['lock_time'][1]-order['lock_time'][0])
```

# 使用分组聚合进行组内计算

```python
import pandas as pd
import numpy as np
```

```python
detail=pd.read_excel('data/meal_order_detail.xlsx')
detail.head()
```

```python
detail_group=detail[['order_id','counts','amounts']].groupby(by='order_id')
```

```python
detail_group.agg(['mean','sum']).head()#求每一列的均值
```

```python
detail_group.agg({'counts':['mean',np.max],'amounts':'std'}).head()#std求标准差
```

```python
#自定义函数
detail_group.agg({'counts':lambda x:sum(x)**2}).head()
```

# 创建透视表与交叉表

```python
import pandas as pd
```

```python
detail=pd.read_excel('data/meal_order_detail.xlsx')
detail.head()
```

```python
pd.pivot_table(detail[['order_id','counts','amounts']],index='order_id',aggfunc='sum').head()
```

```python
pd.pivot_table(detail[['order_id','dishes_name','counts']],index='order_id',columns='dishes_name',aggfunc='sum').head()
pd.pivot_table(detail[['order_id','dishes_name','counts']],index='order_id',columns='dishes_name',values='counts',fill_value=0).head()
```

```python
detail[['order_id','dishes_name','counts']]
pd.crosstab(index=detail['order_id'],columns=detail['dishes_name']).head()#计算订单里有无点这份菜
```

```python
pd.crosstab(index=detail['order_id'],columns=detail['dishes_name'],values=detail['counts'],aggfunc='sum').fillna(0).head()#计算订单里有无点这
```

```python
athletes=pd.read_csv("data/athlete_events.csv")
countries={
    'CHN':'中国',
    'JPN':"日本",
    'KOR':"韩国",
    'USA':'美国',
    'CAN':'加拿大',
    'BRA':'巴西',
    'GBA':'英国',
    'FRA':'法国',
    'ITA':'意大利',
    'ETN':'埃塞尔比亚',
    'KEN':'肯尼亚',
    'NIG':'尼日利亚'
}
my_athletes=athletes[athletes['NOC'].isin(list(countries.keys()))]
```

#### FacetGrid基本使用

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips=sns.load_dataset("tips")
```

```python
g1=sns.FacetGrid(tips)
g1.map(plt.scatter,"total_bill","tip")
```

```python
# g2=sns.FacetGrid(tips,col="day",row="time")
g2=sns.FacetGrid(tips,col="day",hue="time")
g2.map(plt.scatter,"total_bill","tip")
g2.add_legend()
```

```python
g3=sns.FacetGrid(tips,col="smoker",height=4,aspect=0.5)
g3.map(sns.barplot,"day","total_bill",order=['Thur','Fri','Sat','Sun'])
```

```python
# 设置图例
```

```python
g2=sns.FacetGrid(tips,col="day",hue="time")
g2.map(plt.scatter,"total_bill","tip")
# g2.add_legend(title="标题",labels=['午餐','晚餐'],label_order=['Lunch','Dinner']) 
g2.add_legend(title="标题",label_order=['Lunch','Dinner']) 
new_labels=['午餐','晚餐']
for text,label in zip(g2._legend.texts,new_labels):
    text.set_text(label)
plt.rcParams['font.family'] = ['Arial Unicode MS'] #用来正常显示中文标签
sns.set_style('whitegrid',{'font.sans-serif':['Arial Unicode MS','Arial']})
```

```python
g=sns.FacetGrid(tips,col="day",row="time")
g.map(sns.regplot,"total_bill","tip")
g.set_titles(template="时间{row_var}/星期{col_name}")
```

```python
g=sns.FacetGrid(tips,col="day",row="time")
# g.map(sns.regplot,"total_bill","tip")
g.map(sns.scatterplot,"total_bill","tip")
g.set_axis_labels("消费金额","小费")
g.set(facecolor='y')
g.set(xticks=range(0,50,5),xticklabels=["$%d"%x for x in range(0,50,5)])
```

```python
g=sns.FacetGrid(tips,col="day",row="time")
g.map(sns.scatterplot,"total_bill","tip")
g.fig.set_dpi(100)
```
